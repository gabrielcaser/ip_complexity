```python
# Geting countries ids 

# Specify the JSON URL
json_url = "https://oec.world/olap-proxy/data.jsonrecords?cube=complexity_eci_a_hs96_hs6&drilldowns=Country,ECI+Rank,Year&measures=ECI&parents=true&sparse=false"

# Send a GET request to the URL
response = requests.get(json_url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Access the JSON data
    data = response.json()
    # Extract values from the "Country" column
    data = pd.DataFrame(data["data"])
    countries = list(data["Country ID"].unique())
    
    # Print the list of countries
    print(countries)
else:
    # Handle the request error
    print("Error:", response.status_code)
```

```python
import requests

available_years = range(2015, 2024)  # Adjust the range as per the available years

for year in available_years:
  for country in countries:

    # Specify the JSON URL
    json_url = f"https://app-nutria.oec.world/api/stats/relatedness?cube=trade_i_baci_a_92&filter_Country={country}&Year={year}&measures=Trade+Value&parents=true&rca=Exporter+Country,HS4,Trade+Value&alias=Country,HS4&locale=en"

    # Send a GET request to the URL
    response = requests.get(json_url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Access the JSON data
        data = response.json()
        # Process the data as needed
        df = pd.DataFrame(data["data"])
        df["year"] = year
        # Save the DataFrame as a CSV file
        file_path = output_dir + f"/data/{country}_{year}.csv"
        df.to_csv(file_path, index=False)
        print(f"Table downloaded for year {year} and saved as {file_path}")

    else:
        # Handle the request error
        print(f"Error: for {country} in {year}", response.status_code)
```
