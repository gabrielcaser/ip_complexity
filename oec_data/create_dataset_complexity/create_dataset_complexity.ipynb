# Oppening datasets

```python
df_pci_pj = pd.read_csv(work_dir + '/input/OEC_pci_hs4_hs92.csv', sep = ";") # pci at level product p during the year j
```

# Cleaning

```python
# Convert to long format using 'HS4 ID' as the identifier variable
df_pci_pj = pd.melt(df_pci_pj, id_vars=['HS4', 'HS4 ID'], var_name='year', value_name='pci')
```

```python
df_pci_pj = df_pci_pj.sort_values(by=["HS4","year"])
```

```python
df_pci_pj = df_pci_pj.rename(columns = {"HS4": "product_name_hs4", "HS4 ID": "hs4_id"})
```

# Saving 

```python
df_pci_pj["year"] = df_pci_pj["year"].astype(int)
```

```python
df_pci_pj.to_csv(output_dir + '/data/complexity_dataset.csv', index = False)
```
